```python
from googletrans import Translator
```

```python
translate = Translator()
mysent = '众多拉美特色产品共赴“进博之约” 外交部：中国市场，世界共享外交部发言人毛宁6日在例行记者会上表示，随着中拉高质量共建“一带一路”深入推进，拉美和加勒比国家将从中国市场中发现更多机遇，获得更大发展'
```

```python
mysent
```

```python
hindi = translate.translate(mysent, src = 'zh-cn', dest = 'hi')
print('Original_text = ',mysent)
print('Translater_text =',hindi.text)
```

## Hindi

```python
from deep_translator import GoogleTranslator

mysent = '众多拉美特色产品共赴“进博之约” 外交部：中国市场，世界共享外交部发言人毛宁6日在例行记者会上表示，随着中拉高质量共建“一带一路”深入推进，拉美和加勒比国家将从中国市场中发现更多机遇，获得更大发展'
hindi = GoogleTranslator(source='zh-CN', target='hi').translate(mysent)

print('Original text:', mysent)
print('Translated text:', hindi)
```

```python
mysent = 'வேட்டையன் ஓடிடி ரிலீஸ் தேதி உறுதியானது.. இந்த வீக் எண்ட்டை கொண்டாட ரசிகர்களே ரெடியா?தொலைக்காட்சி'
myspeech = gTTS(mysent, lang = 'hi')
myspeech.save('hindi.mp3')
```

```python
import fitz
import io
import PIL.Image
```

```python
pdf = fitz.open(r"C:\Users\DeLL\Documents\Data Science\Deep Learning\Deep Learning by Sundaram\NLP\Pdf for info extraction\Painting.pdf")
counter = 1
for i in range(len(pdf)):
    page = pdf[i]
    images = page.get_images()
    for image in images:
        base_img = pdf.extract_image(image[0])
        image_data = base_img['image']
        extension = base_img['ext']
        img = PIL.Image.open(io.BytesIO(image_data))
        img.save(open(f'{counter}.{extension}','wb'))
        counter += 1
```
