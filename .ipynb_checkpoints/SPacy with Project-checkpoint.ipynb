```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
## BY default this package has the capability to identify which sentence is positive or negative
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer(text):
    doc = nlp(text)
    sentiment = analyzer.polarity_scores(text)
    sentiment_score = sentiment['compound']
    if sentiment_score >= 0.05:
        sentiment_label = 'Positive'
    elif sentiment_score <= -0.05:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    return{'Text' : text, 'Token' : [token.text for token in doc], 'sentiment_score' : sentiment_score, 'Sentiment_label' : sentiment_label}
```

```python
# Polarity score gives the threshold values
```

```python
texts = [
    'NLP is one of the hot topics now a days',
    'there are lots of demand in market about Sentiment Analysis',
    'I am not sure how I feel about the NLP',
    'I like this product very much',
    "I don't like this product at all",
    'This is worst experience i have ever had']
```

```python
for text in texts:
    result = sentiment_analyzer(text)
    print(result)
```

```python
text = 'My Workdesk is an organized mess'
sentiment_analyzer(text)
```
