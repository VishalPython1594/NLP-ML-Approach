# Images from pdf

```python
import fitz
import io
import PIL.Image

pdf = fitz.open(r"C:\Users\DeLL\Documents\Data Science\Deep Learning\Deep Learning by Sundaram\NLP\Pdf for info extraction\Artificial_intelligence.pdf")

counter = 100
for i in range(len(pdf)):
    page = pdf[i]
    images = page.get_images()
    for image in images:
        base_img = pdf.extract_image(image[0])
        image_data = base_img['image']
        extension = base_img['ext']
        img = PIL.Image.open(io.BytesIO(image_data))
        img.save(open(f'image{counter}.{extension}', 'wb'))
        counter += 1
```

### SPeech to text

```python
from deep_translator import GoogleTranslator
from playsound import playsound
from gtts import gTTS
import speech_recognition as sr
recog = sr.Recognizer()

with sr.Microphone() as source:
    print('Please Speak')
    audio = recog.listen(source)

try:
    text = recog.recognize_google(audio)
    print('You Said.... ', text)

except sr.UnknownValueError:
    print('Voice not audible')

except:
    print('still no luck')

res = GoogleTranslator(source = 'hi', target = 'bn').translate(text)
res = gTTS(res, lang = 'bn')
audio_file = '15.11.mp3'
res.save(audio_file)
playsound(audio_file)
```

```python
from nltk.corpus import movie_reviews
```

```python
movie_reviews.fileids()
```

```python
wordlis = movie_reviews.words('neg/cv016_4348.txt')
```

```python
entire_text = ' '.join(wordlis)
entire_text
```

```python
clean_paragraph = []
sentence = nltk.tokenize.sent_tokenize(entire_text)
for i in range(len(sentence)):
    sent = re.sub(r'[^a-zA-Z]', ' ', sentence[i])
    sent = sent.lower()
    sent = sent.split()
    sent = [lemmatization.lemmatize(word) for word in sent if not word in set(stopwords.words('english'))]
    sent = ' '.join(sent)
    clean_paragraph.append(sent)

clean_paragraph
```
